```python
# Continue in notebooks/01_data_simulation_and_eda.ipynb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv("../data/simulated_sensor_data.csv") # Adjusted path for notebook
```

```python
print("DataFrame Head:")
print(df.head())

print("\nDataFrame Description:")
print(df.describe())

print("\nAnomaly Value Counts:")
print(df['anomaly'].value_counts())
```

```python
# Ensure 'timestamp' is a datetime object
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
# Time series plot for temperature and pressure
plt.figure(figsize=(15, 6))
plt.plot(df['timestamp'], df['temperature'], label='Temperature')
plt.plot(df['timestamp'], df['pressure'], label='Pressure')
plt.title("Temperature and Pressure over Time")
plt.xlabel("Timestamp")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
# Box plots to visualize distribution and potential outliers
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df['temperature'])
plt.title("Box Plot of Temperature")

plt.subplot(1, 2, 2)
sns.boxplot(y=df['pressure'])
plt.title("Box Plot of Pressure")
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
print("\nCorrelation Matrix:")
print(df[['temperature', 'pressure', 'anomaly']].corr())

plt.figure(figsize=(6, 5))
sns.heatmap(df[['temperature', 'pressure', 'anomaly']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Features")
plt.show()
```
